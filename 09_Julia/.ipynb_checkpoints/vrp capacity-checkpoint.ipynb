```python
import numpy as np
import random
import seaborn as sns
import matplotlib.pyplot as plt
from tqdm import tqdm 
class Colony:
    def __init__(self,size):
        self.ants=[]
        self.size = size
        
    def __repr__(self):
        return f'{self.ants}'
class Ant:
    def __init__(self,tour,fitness):
        self.tour = tour
        self.fitness=fitness
    
    def __repr__(self):
        return f' tour: {self.tour} , fitness: {self.fitness}'


def roulette_wheel(prob):
    return np.random.choice(len(prob),p=prob)
    

def update_phromone(tau,colony):

    for i in range(colony.size):
        ant=colony.ants[i]

        for k in range(len(ant.tour)):
            tour=ant.tour[k]
            for j in range(1,len(tour)):
                tau[tour[j-1]][tour[j]] =  tau[tour[j-1]][tour[j]]  + 1/ ant.fitness
                tau[tour[j]][tour[j-1]] =  tau[tour[j]][tour[j-1]]  + 1/ ant.fitness
  
```

```python
      
def createColony( graph, colony , antNo, tau, eta, alpha,  beta, hub,capacities,max_capacity):
    node_len = len(graph)

    colony=Colony(antNo)
    for i in range(antNo):
        tours = create_one_ant(graph,tau,eta,alpha,beta,hub,capacities,max_capacity)
        ant = Ant(tours.copy(),obj_function(tours.copy(),graph))
        colony.ants.append(ant)
    return colony
        
```

```python
       
def obj_function(routes,graph):
    
    fitness=0
    for i in range(len(routes)):
        for j in range(len(routes[i])-1):
            fitness+=graph[routes[i][j]][routes[i][j+1]]
        fitness+= graph[routes[i][0]][routes[i][-1]]
    
    return fitness
```

```python
graph,nodes= create_graph_new()
```

```python
def create_one_ant( graph,  tau, eta, alpha,  beta, hub,capacity,max_capacity):
    vertices=[i for i in range(len(graph)) if i!=hub]
    routes=[]
    visited=[hub]
    while len(vertices)!=0:
        current_node = np.random.choice(vertices)
        
        current_capacity=max_capacity-capacity[current_node]
        vertices.remove(current_node)
        route =[hub]
        visited.append(current_node)
        route.append(current_node)
        
        while len(vertices)!=0 :
            p_all_nodes =  tau[current_node,:]**alpha * eta[current_node,:]**beta
            p_all_nodes[visited]=0
            capacity_nodes=capacity.copy()
            capacity_nodes[capacity_nodes>current_capacity]=0
#             print(current_capacity,capacity_nodes)
            p_all_nodes=p_all_nodes*capacity_nodes
            
            if np.sum(p_all_nodes)==0:
                break;
            prob= p_all_nodes/np.sum(p_all_nodes)
            current_node=roulette_wheel(prob)
            visited.append(current_node)
            vertices.remove(current_node)
            current_capacity=current_capacity-capacity[current_node]
            route.append(current_node)
        route.append(hub)
        routes.append(route.copy())
    
    return routes
            
            
```

```python
maxIter = 1000
antNo = 50




tau_initial = 10 * 1 / (len(nodes) *  np.mean(graph))

tau = np.ones((len(graph),len(graph)))

eta = 1 / graph
max_rho=0.6

min_rho=0.05
rho=0.05
alpha=1
beta =4
hub=0
capacity = np.array([0,1100,700,800,1400,2100,400,800,100, 500, 600, 1200, 1300, 1300, 300, 900, 2100, 1000, 900, 2500, 1800, 700])
max_capacity = 6000
best_fitness=np.inf
best_tour=[]
values=[]
max_values=[]
each_iter_result=[]
for i in tqdm(range(maxIter)):
    colony=[]
    colony = createColony( graph, colony , antNo, tau, eta, alpha,  beta,hub,capacity,max_capacity)
    best_ant_in_colony = sorted(colony.ants,key=lambda x: x.fitness)[0]
    if best_fitness> best_ant_in_colony.fitness:
        best_fitness= best_ant_in_colony.fitness
        best_tour=best_ant_in_colony.tour

    update_phromone(tau,colony)
    values.append(best_fitness)
    max_values.append(sorted(colony.ants,key=lambda x: x.fitness,reverse=True)[0].fitness)
    each_iter_result.append(best_ant_in_colony.fitness)
    
    rh0 = max_rho - i * ( (max_rho - min_rho) / (maxIter))
    tau  = ( 1 - rho ) * tau



print(best_fitness,best_tour)
plt.figure(figsize=(20,7))

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
plt.subplot(1, 2, 1)
for i in range(len(best_tour)):
    x_corrdinates=[nodes[i][0] for i in best_tour[i]]
    y_coordinates=[nodes[i][1] for i in best_tour[i]]
    plt.plot(x_corrdinates,y_coordinates,marker='o',color=(np.random.rand(),np.random.rand(),np.random.rand()))
plt.subplot(1, 2, 2)

sns.lineplot(range(maxIter),values,label="best solution till that iteration")
sns.lineplot(range(maxIter),each_iter_result,label="best solution within that colony")  
sns.lineplot(range(maxIter),max_values,label="worst solution within that colony")  

plt.show()
    
```

```python
def create_one_ant_saving( graph,  tau, eta, alpha,  beta, hub,capacity,max_capacity,saving,gamma):
    vertices=[i for i in range(len(graph)) if i!=hub]
    routes=[]
    visited=[hub]
    while len(vertices)!=0:
        
        current_node = np.random.choice(vertices)
        
        current_capacity=max_capacity-capacity[current_node]
        vertices.remove(current_node)
        route =[hub]
        visited.append(current_node)
        route.append(current_node)
        
        while len(vertices)!=0 :
            p_all_nodes =  tau[current_node,:]**alpha * eta[current_node,:]**beta * saving[current_node,:]**gamma
            p_all_nodes[visited]=0
            capacity_nodes=capacity.copy()
            capacity_nodes[capacity_nodes>current_capacity]=0
#             print(current_capacity,capacity_nodes)
            p_all_nodes=p_all_nodes*capacity_nodes
            
            if np.sum(p_all_nodes)==0:
                break;
            prob= p_all_nodes/np.sum(p_all_nodes)
            current_node=roulette_wheel(prob)
            visited.append(current_node)
            vertices.remove(current_node)
            current_capacity=current_capacity-capacity[current_node]
            route.append(current_node)
        route.append(hub)
        routes.append(route.copy())
    
    return routes
            

def createColony_saving( graph, colony , antNo, tau, eta, alpha,  beta, hub,capacities,max_capacity,saving,gamma):
    node_len = len(graph)

    colony=Colony(antNo)
    for i in range(antNo):
        tours = create_one_ant_saving(graph,tau,eta,alpha,beta,hub,capacities,max_capacity,saving,gamma)
        ant = Ant(tours.copy(),obj_function(tours.copy(),graph))
        colony.ants.append(ant)
    return colony
        
# with saving matrix
maxIter = 100
antNo = 50
saving_matrix=create_saving_matrix(graph,0)



tau_initial = 10 * 1 / (len(nodes) *  np.mean(graph))

tau = np.ones((len(graph),len(graph)))

eta = 1 / graph
max_rho=0.6

min_rho=0.05
rho=0.05
alpha=1
beta =4
hub=0
capacity = np.array([0,1100,700,800,1400,2100,400,800,100, 500, 600, 1200, 1300, 1300, 300, 900, 2100, 1000, 900, 2500, 1800, 700])
max_capacity = 6000
best_fitness=np.inf
best_tour=[]
values=[]
max_values=[]
each_iter_result=[]
for i in tqdm(range(maxIter)):
    colony=[]
    colony = createColony_saving( graph, colony , antNo, tau, eta, alpha,  beta,hub,capacity,max_capacity,saving_matrix,2)
    best_ant_in_colony = sorted(colony.ants,key=lambda x: x.fitness)[0]
    if best_fitness> best_ant_in_colony.fitness:
        best_fitness= best_ant_in_colony.fitness
        best_tour=best_ant_in_colony.tour

    update_phromone(tau,colony)
    values.append(best_fitness)
    max_values.append(sorted(colony.ants,key=lambda x: x.fitness,reverse=True)[0].fitness)
    each_iter_result.append(best_ant_in_colony.fitness)
    
    rh0 = max_rho - i * ( (max_rho - min_rho) / (maxIter))
    tau  = ( 1 - rho ) * tau



print(best_fitness,best_tour)
plt.figure(figsize=(20,7))

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("graph")
plt.subplot(1, 2, 1)
for i in range(len(best_tour)):
    x_corrdinates=[nodes[i][0] for i in best_tour[i]]
    y_coordinates=[nodes[i][1] for i in best_tour[i]]
    plt.plot(x_corrdinates,y_coordinates,marker='o',color=(np.random.rand(),np.random.rand(),np.random.rand()))
plt.subplot(1, 2, 2)

sns.lineplot(range(maxIter),values,label="best solution till that iteration")
sns.lineplot(range(maxIter),each_iter_result,label="best solution within that colony")  
sns.lineplot(range(maxIter),max_values,label="worst solution within that colony")  

plt.show()
    
```
