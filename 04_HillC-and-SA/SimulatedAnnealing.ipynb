# Simulated Annealing

```python
import numpy as np
import matplotlib.pyplot as plt
import random
```

## Single state, contains position and cost

```python
class State:
    def __init__(self) -> None:
        self.position = []
        self.cost = 0

    def __repr__(self) -> str:
        return f'Node(position={self.position}, cost={self.cost})'
```

## Make a plot for iteration vs cost

```python
def make_plot(costs):
  fig = plt.figure()
  ax = fig.add_subplot(1, 1, 1)
  ax.set_ylabel('Value of A.cost')
  ax.set_xlabel('Iteration #')
  ax.plot(costs)
```

## Objective function

```python
def objective(X, func_no):
    if func_no == 1:
        return 10 * -np.cos(2 * (X[0] ** 2+X[1] ** 2) ** 0.5) * \
                        np.exp(-0.5 * ((X[0]+1) ** 2+(X[1]-1) ** 2) ** 0.5) + 5.1
    if func_no == 2:
        return -1*(0.2 + X[0]**2 + X[1]**2 - 0.1*np.cos(6*np.pi*X[0]) - 0.1*np.cos(6*np.pi*X[1]) )
    if func_no == 3:
        return -1 * (X[0]**2 + X[1]**2)
```

## Defining the neighbour funtion

```python
def find_neighbour(A, step_size, ub, lb, func_no):
    neighbours = []
    n_variables = len(A.position)

    for i in range(n_variables):

        B = State()
        B.position = A.position.copy()
        B.position[i] = min(ub[i], B.position[i] + step_size) 
        B.cost = objective(B.position, func_no)
        neighbours.append(B)

        B = State()
        B.position = A.position.copy()
        B.position[i] = max(lb[i], B.position[i] - step_size)
        B.cost = objective(B.position, func_no)
        neighbours.append(B)
        
    return random.choice(neighbours)
```

## Simulated annealing funtion

```python
def find_optimum(A, step_size, ub, lb, func_no, max_iter, T, alpha):

    costs = []

    for  i in range(max_iter):
        costs.append(A.cost)
        B = find_neighbour(A, step_size, ub, lb, func_no)
        delta = A.cost - B.cost

        if delta < 0:
            A.position = B.position.copy()
            A.cost = B.cost
        else:
            p = np.exp(-delta / T);
            r = np.random.rand(1)
            if r < p:
                A.position = B.position.copy()
                A.cost = B.cost
    
        T = alpha * T
    
    make_plot(costs)
    return A
```

```python
A = State()
func_no = 1
A.position = np.array([0,0], dtype=float)
A.cost = objective(A.position, func_no)
step_size = 0.001
ub = np.array([2, 2], dtype=float)
lb = np.array([-2, -2], dtype=float)
max_iter = 30000
T = 1
alpha = 0.999

find_optimum(A, step_size, ub, lb, func_no, max_iter, T, alpha)
```
